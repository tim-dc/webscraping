### Importing

```python
import json
import pandas as pd
from newsapi import NewsApiClient

# initialization
newsapi = NewsApiClient(api_key='350ea48cbdb44b44a414096f386c8e29')

date_from = '2022-04-21'
date_to = '2022-04-22'

articles = newsapi.get_everything(sources = 'bbc-news',
                                  from_param =  date_from,
                                  to = date_to,
                                  language = 'en',
                                  sort_by = 'relevancy',
                                  page=1)

with open('newsjson.txt', 'w') as outfile:
    json.dump(articles, outfile, indent=3)
```
